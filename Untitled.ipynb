## 평균(mean)으로 보간
6.0482555379

```python
fill_mean = data.copy()
fill_mean['Value'].fillna(fill_mean.Value.mean(), inplace=True)
print(f"Mean : {fill_mean['Value'].mean()}")
fill_mean
```

```python
plt.figure(figsize=(12, 4))


plt.plot(data['Value'], color='blue', label='Original Data')
plt.plot(fill_mean['Value'].where(data['Value'].isnull()), color='red', label='Fill Mean')

plt.title("Filled Mean")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
fill_mean.describe()
```

```python
fill_mean.to_csv('./data/submission_fill_mean.csv', index=False)
```

## 중앙값(median)으로 보간

## 최빈값(mode)으로 보간
6.2616609164	

```python
fill_mode = data.copy()
fill_mode['Value'].fillna(fill_mode['Value'].mode(dropna=True)[0], inplace=True)
print(f"Mode : {fill_mode['Value'].mode()}")
fill_mode
```

```python
plt.figure(figsize=(12, 4))


plt.plot(data['Value'], color='blue', label='Original Data')
plt.plot(fill_mode['Value'].where(data['Value'].isnull()), color='red', label='Fill Mode')

plt.title("Filled Mode")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
fill_mode.to_csv('./data/submission/submission_fill_mode.csv', index=False)
```

## spline으로 보간
3.0326187835

```python
fill_spline = data.copy()
fill_spline['Value'] = fill_spline['Value'].interpolate(method='spline', order=3)
```

```python
plt.figure(figsize=(12, 4))


plt.plot(data['Value'], color='blue', label='Original Data')
plt.plot(fill_spline['Value'].where(data['Value'].isnull()), color='red', label='Fill Spline')

plt.title("Fill Spline")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
fill_spline
```

```python
fill_spline.to_csv('./data/submission/submission_fill_spline.csv', index=False)
```

## rolling window로 보간 
- test 1 : window= 5000, min_peroid=1 -> 4.4533853973	

```python
rolling_mean = data['Value'].rolling(window=5000, min_periods=1).mean()
```

```python
fill_rolling_mean = data.copy()
fill_rolling_mean['Value'] = fill_rolling_mean['Value'].fillna(rolling_mean)
```

```python
plt.figure(figsize=(12, 4))


plt.plot(data['Value'], color='blue', label='Original Data')
plt.plot(fill_rolling_mean['Value'].where(data['Value'].isnull()), color='red', label='Fill Rolling Window(window=5000, min_peroid=1)')

plt.title("Fill Rolling Window")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
fill_rolling_mean.to_csv('./data/submission/submission_fill_rolling_mean-5000-1.csv', index=False)
```
